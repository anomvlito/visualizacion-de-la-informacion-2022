```python
def primero(row):
    # 0 es E 1 es I 
    if row.mbti[0] == "E":
        return 0
    elif row.mbti[0] == "I":
        return 1
def segundo(row):
    # 0 es S 1 es N 
    if row.mbti[1] == "S":
        return 0
    elif row.mbti[1] == "N":
        return 1
def tercero(row):
    # 0 es F 1 es T 
    if row.mbti[2] == "F":
        return 0
    elif row.mbti[2] == "T":
        return 1
def cuarto(row):
# 0 es P 1 es J 
    if row.mbti[3] == "P":
        return 0
    elif row.mbti[3] == "J":
        return 1
```

```python

def grupo(row):
    if row.mbti in ["INTJ","INTP", "ENTJ", "ENTP"]:
        return "Analista"
    elif row.mbti in ["INFJ","INFP", "ENFJ", "ENFP"]:
        return "Diplomatico"
    elif row.mbti in ["ISTJ","ISFJ", "ESTJ", "ESFJ"]:
        return "Sentinela"
    elif row.mbti in ["ISTP","ISFP", "ESTP", "ESFP"]:
        return "Explorador"
    else:
        return "ya era"
```

```python
def mbti(row):
    return row.personality_type[0:4]

def eneagrama(row):
    return row.personality_type[4:]
def lencito(row):
    return len(row.mbti)





copia["mbti"] = copia.apply(mbti,axis =1)
copia["eneagrama"] = copia.apply(eneagrama,axis =1)


copia.head(70)
```

```python

copia["grupo"] = copia.apply(grupo, axis = 1)

    
```

```python
copia.head(40)
```

```python


filtro = copia[(copia["category_name"] == 'Anime') & (copia["mbti"] != 'XXXX') ]


print(filtro.shape)
filtro.head(5)

```

```python
df = filtro.drop(["personality_type", "category_id", "category_name"], axis=1)
```

```python
df.head()
```

```python
nombres_anime = set()

for index in df.index:
    nombres_anime.add(df["subcategory_name"][index])
    
print(len(nombres_anime))
```

```python
dic_animes_agrupados = {}

for anime in nombres_anime:
    dic_animes_agrupados[anime] = df.loc[df["subcategory_name"] == anime]
            
            
```

```python
for dic in dic_animes_agrupados:
    print(dic)
```

```python
lista_jsons_anime = []

for dic in dic_animes_agrupados:
    childrencito = []
    for k in range(len(dic_animes_agrupados[dic])):
        childrencito.append({
            "name" : dic_animes_agrupados[dic].iloc[k].character_name,
            "grupo": dic_animes_agrupados[dic].iloc[k].grupo,
            "mbti": dic_animes_agrupados[dic].iloc[k].mbti,
            "eneagrama" : dic_animes_agrupados[dic].iloc[k].eneagrama,
            "votos" : int(dic_animes_agrupados[dic].iloc[k].vote_count)
        })
    lista_jsons_anime.append({"name": dic,"children": childrencito})
        
    #print(dic_animes_agrupados[dic].iloc[0].mbti)
    
```

```python
dic_final = {"children": lista_jsons_anime}
```
